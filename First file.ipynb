```python
sns.countplot(x='Gender',data=df) #count the gender and display through graph
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)     #this for loop represnt through the number 
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) #describe the amont used by genders
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) #describe the amont used by genders
ax=sns.barplot(x='Gender',y='Amount',data=sales_gen)  #represent in graph
for bars in ax.containers:
    ax.bar_label(bars)    #for loop used for exact amount used by  gender
```

```python
STATE
```

```python
#total number of orders from top 10 states 
sales_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(17,5)}) #here fix the size
sns.barplot(data=sales_states,x='State',y='Orders')
```

```python
#total amount/sales from top 10 states
sales_states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_states,x='State',y='Amount')
```

```python
MARITAL STATUS
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=sales_states,x='Marital_Status',y='Amount',hue='Gender')
```

```python
OCCUPATION 
```

```python
#count all the occupation buys the crackers
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(20,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#based on amount which occupation buys more(purcahsing power)
sales_states=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_states,x='Occupation',y='Amount')
```

```python
PRODUCT CATEGORY
```

```python
#count all the occupation buys the crackers   (count of order)
ax=sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(22,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 #based on amount which product_category  buys more(purcahsing power)
sales_states=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_states,x='Product_Category',y='Amount')
```
